5CV

```python
from sklearn.model_selection import StratifiedKFold


skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

folds = []

for i, (train_idx, val_idx) in enumerate(skf.split(df,df_label['label'])):
    train_data = df.iloc[train_idx, :]
    val_data = df.iloc[val_idx, :]

    y_train = df_label['label'].iloc[train_idx]
    y_val = df_label['label'].iloc[val_idx]

    fold_data = {
        'X_train': train_data,
        'X_val': val_data,
        'y_train': y_train,
        'y_val': y_val
    }

    folds.append(fold_data)
```

```python
unique_values, counts = np.unique(folds[0]['y_train'], return_counts=True)
display(dict(zip(unique_values, counts)),np.max(counts))
```

```python
import rpy2.rinterface_lib.callbacks
import logging
# Ignore R warning messages
#Note: this can be commented out to get more verbose R output
rpy2.rinterface_lib.callbacks.logger.setLevel(logging.ERROR)

%load_ext rpy2.ipython
```

```python
from rpy2.robjects import conversion, default_converter
```

```python
with conversion.localconverter(default_converter):
  %R install.packages("BiocManager")
  %R BiocManager::install('scran')
  %R library("scran")
```

```python
with conversion.localconverter(default_converter):
  %R install.packages("SingleCellExperiment")
  %R library("SingleCellExperiment")
```

```python
hvg_folds = [] # run once
```

run for each split

```python
X_train = folds[0]['X_train']
X_val = folds[0]['X_val']

y_train = folds[0]['y_train']
y_val = folds[0]['y_val']

X_train_T = X_train.T
```

```python
%R -i X_train_T
%R sce <- SingleCellExperiment(assays = list(logcounts=as.matrix(X_train_T)))

%R dec <- modelGeneVar(sce)
%R top_hvgs <- getTopHVGs(dec,n=3000) #prop=0.1 default %10 top

%R -o top_hvgs
```

```python
fold_data = {
            'X_train': X_train[top_hvgs],
            'X_val': X_val[top_hvgs],
            'y_train': y_train,
            'y_val': y_val
        }

hvg_folds.append(fold_data)
```
