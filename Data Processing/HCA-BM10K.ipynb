Subset Data

```python
%%R
set.seed(1)

target_n <- 10000L
cell_type <- colLabels(sce.bone)
```

```python
%%R
# counts, proportions, and ideal sample sizes
tab <- table(cell_type)
n_float <- as.numeric(prop.table(tab) * target_n)
n_floor <- floor(n_float)
remainder <- target_n - sum(n_floor)
```

```python
%%R
# distribute the leftover 1-by-1 to the largest fractional parts
frac <- n_float - n_floor
add1 <- order(frac, decreasing = TRUE)[seq_len(remainder)]
n_take <- n_floor
n_take[add1] <- n_take[add1] + 1L
names(n_take) <- names(tab)
```

```python
%%R
# sample per class
idx <- unlist(lapply(names(n_take), function(ct) {
  pool <- which(cell_type == ct)
  # guard in case target exceeds available for a rare type
  size <- min(n_take[[ct]], length(pool))
  sample(pool, size = size)
}), use.names = FALSE)
```

```python
%%R
sce.bone_10k <- sce.bone[, idx]
```
