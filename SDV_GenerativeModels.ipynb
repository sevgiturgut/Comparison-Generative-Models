```python
import pickle
# Load object from file
with open(gdrivePath +os.sep + "data"+os.sep+'X_train.pkl', 'rb') as f:
    X_train = pickle.load(f)
with open(gdrivePath +os.sep + "data" +os.sep+'X_test.pkl', 'rb') as f:
    X_test = pickle.load(f)
with open(gdrivePath +os.sep + "data"+os.sep+'y_train.pkl', 'rb') as f:
    y_train = pickle.load(f)
with open(gdrivePath +os.sep + "data"+os.sep+'y_test.pkl', 'rb') as f:
    y_test = pickle.load(f)
```

```python
print(X_train.shape,y_train.shape, X_test.shape,y_test.shape)
```

Metadata

```python
type(X_train)
```

```python
X_train_pd = pd.DataFrame(X_train)
X_train_pd['class'] = y_train
X_train_pd
```

```python
metadata = SingleTableMetadata()
metadata.detect_from_dataframe(X_train_pd)
```

```python
unique_values, counts = np.unique(y_train, return_counts=True)
display(dict(zip(unique_values, counts)),np.max(counts))
```

```python
unique_values, counts = np.unique(y_train, return_counts=True)
display(dict(zip(unique_values, counts)))
```

```python
threshold = 759

# Create a dictionary for samples_per_class
samples_per_class = {}

# Loop through the counts and calculate the number of samples to generate
for class_id, (count_index, count) in zip(unique_values, enumerate(counts)):
    # If count is less than the threshold, calculate how many samples to generate
    if count < threshold:
        samples_per_class[class_id] = threshold - count
    else:
        samples_per_class[class_id] = 0  # No samples needed for this class

# Print the result
print("Samples Per Class:", samples_per_class)
```

Sampling

ctGAN

```python
ctgan = CTGANSynthesizer(metadata,epochs=50,verbose=False)
ctgan.fit(X_train_pd)
```

```python
ctgan_synthetic = pd.DataFrame(columns=X_train_pd.columns)
y_synthetic = []
for class_id, num_samples in samples_per_class.items():

  condition = cd(column_values={'class':class_id},num_rows=num_samples)

  synthetic = ctgan.sample_from_conditions(conditions=[condition])
  ctgan_synthetic = pd.concat([ctgan_synthetic, synthetic], axis=0)
  y_synthetic = y_synthetic + [class_id]*num_samples
y_synthetic = np.array(y_synthetic)
```

TVAE

```python
tvae = TVAESynthesizer(metadata,epochs=50)
tvae.fit(X_train_pd)
```

```python
tvae_synthetic = pd.DataFrame(columns=X_train_pd.columns)
for class_id, num_samples in samples_per_class.items():

  condition = cd(column_values={'class':class_id},num_rows=num_samples)

  synthetic = tvae.sample_from_conditions(conditions=[condition])
  tvae_synthetic = pd.concat([tvae_synthetic, synthetic], axis=0)

```

GC

```python
gc = GaussianCopulaSynthesizer(metadata)
gc.fit(X_train_pd)
```

```python
gc_synthetic = pd.DataFrame(columns=X_train_pd.columns)
for class_id, num_samples in samples_per_class.items():

  condition = cd(column_values={'class':class_id},num_rows=num_samples)

  synthetic = gc.sample_from_conditions(conditions=[condition])
  gc_synthetic = pd.concat([gc_synthetic, synthetic], axis=0)

```

Combine original train and synthetic samples

```python
X_train_gen = pd.concat(([X_train_pd, ctgan_synthetic], axis=0) #tvae_synthetic or gc_synthetic
y_train_gen = np.concatenate(([y_train, y_synthetic], axis=0)
y_train_indexes = np.concatenate(([np.full(len(y_train), 1), np.full(len(y_synthetic), 2)], axis=0)
```
